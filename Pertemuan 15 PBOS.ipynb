```python
lagu = []
def show_data():
    if len(lagu)<= 0:
        print ("belum ada data")
       
    else: 
        for indeks in range (len(lagu)):
            print ("[%d] %s " % (indeks, lagu[indeks]))
            

def insert_data():           
    lagu_baru = input("judul lagu : ")
    lagu.append(lagu_baru)
    show_menu()

def edit_data():
    show_data()
    indeks = eval(input("inputkan no lagu : "))
    print(len(lagu))
    if (indeks > len(lagu)):
        print("no lagu salah")
        show_menu()
        
    else:
        judul_baru = input("judul lagu : ")
        lagu[indeks] = judul_baru
        show_menu()
        
        
def delete_data():
    show_data()
    indeks = eval(input("inputkan no lagu: "))
    if(indeks > len(lagu)):
        print ("no lagu salah")
        show_menu()
    
       
    else :
        lagu.remove(lagu[indeks])
        show_menu()
        
def show_menu():
    print ("\n")
    print ("===========menu lagu ============")
    print ("[1] show data ")
    print ("[2] insert data")
    print ("[3] edit data")
    print ("[4] delet data")
    print ("[5] exit ")
    
    menu = input("pilih menu> ")
    print ("\n")
    if menu == "1":
        show_data()
    elif menu == "2":
        insert_data()
    elif menu == "3":
        edit_data()
    elif menu == "4":
        delete_data()
    elif menu == "5":
        exit()
    else:
        print("salah pilih!")
        
show_menu()
```
